```python
import random

df1_applied = df1.copy()

def averageScore(row):
    teachingScore = row['Teaching Score']
    researchScore = row['Research Score']
    citationScore = row['Citations Score']
    industryIncomeScore = row['Industry Income Score']
    internationalOutlookScore = row['International Outlook Score']
    averageScore = (teachingScore + researchScore + citationScore + industryIncomeScore + internationalOutlookScore) // 5
    return averageScore

df1_applied['Average Score'] = df1_applied.apply(averageScore, axis=1)
df1_applied
```

```python
from sklearn.impute import SimpleImputer

df1_copy = df1.copy()

imputer = SimpleImputer(strategy='mean')

columns_to_input = ['Teaching Score', 'Research Score', 'Citations Score', 'Industry Income Score', 'International Outlook Score']

df1_copy[columns_to_input] = df1_copy[columns_to_input].applymap(lambda x: float(str(x).replace(',', '')))
df1_copy[columns_to_input] = imputer.fit_transform(df1_copy[columns_to_input])

df1_copy[columns_to_input] = df1_copy[columns_to_input].applymap(lambda x: f"{x:,.0f}")

df1_copy
```
